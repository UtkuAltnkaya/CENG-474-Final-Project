```python

def fill_missing(i,col,condition):
     index = []
     j = i
     valid = ''
     while (j != i + 8):
        #  pd.isna(data.at[j,col])
         if condition(j,col):
             index.append(j)
         else:
             valid = data.at[j,col]
         j+=1
     for k in index:
         data.at[k,col] = valid

```
